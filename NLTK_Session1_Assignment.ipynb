```python
import wikipediaapi
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
wiki_wiki = wikipediaapi.Wikipedia('NLTKProject (vandhanavishnuraj09@gmail.com)', 'en')
```

```python
page_py = wiki_wiki.page("India")
```

```python
if page_py.exists():
  print(page_py.text)
else:
  print("Page doesn't exist")
```

```python
article = page_py.text
```

```python
article
```

```python
article_tokenize = word_tokenize(article)
```

```python
article_tokenize
```

```python
ensw = stopwords.words('english')
```

```python
article_tokenized = [x.lower() for x in article_tokenize]
```

```python
filtered_article = [item for item in article_tokenized if item.isalpha() and item not in ensw]
```

```python
filtered_article
```

```python
pstem = PorterStemmer()
```

```python
#doubt - why not stemmed_article = pstem.stem(filtered_article) instead of the one below
stemmed_article = [pstem.stem(item) for item in filtered_article]
```

```python
stemmed_article
```

```python
lzr = WordNetLemmatizer()
```

```python
lemmatized_article = [lzr.lemmatize(item) for item in stemmed_article]
```

```python
#doubt - parts of speech mentioning is necessary? pos='v'
lemmatized_article
```

```python
freq_words = [lemmatized_article.count(w) for w in lemmatized_article]
freq_count = dict(list(zip(lemmatized_article, freq_words)))
```

```python
freq_count
```

```python
sorted_freq_words = sorted(freq_count.items(), key = lambda item: item[1], reverse=True)
```

```python
sorted_freq_words
```
